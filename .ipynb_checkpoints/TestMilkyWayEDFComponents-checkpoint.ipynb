# Test Milky Way EDF components

```python
%matplotlib inline

import numpy as np
import agama
import matplotlib.pyplot as plt
import ChemicalDistributions as cd
import StarFormationHistories as sfh
import PhaseSpaceDistributions as psd
```

## Test star formation law

```python
# Modify star formation parameters
taumax    = 9.5#8.#11
sfrdecay  = 5.#12.#1.#9
sfrgrowth = 1.

# Create Milky Way EDF instance 
agemin  = 0.
agemax  = 13.1
nage    = 100
agegrid = np.linspace(agemin,agemax,nage)
sfr     = sfh.StarFormationRateDoubleExponential(agegrid,taumax,sfrdecay,sfrgrowth) 

# Create plot
plt.rc('font',family='serif') 
fig    = plt.figure(figsize=(6,4))
plt.plot(agegrid,sfr,'-k',linewidth=2.0) 
plt.xlabel(r"$\tau (Gyr)$",fontsize=16)
plt.ylabel("Star formation rate",fontsize=16)
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12)
fig.savefig("../plots/edf/sfr.eps",format='eps',bbox_inches='tight')
```

### Test [Fe/H] profile

```python
# Modify [Fe/H] parameters
fehbirth     = -1.0
tauenrich    = 3.0
fehLzsolgrad = -0.000256
Lzsol        = 2076.
fehsig       = 0.2

# Calculate probabilities
feh     = -1.0
nfeh    = 100
fehgrid = np.zeros(nfeh)+feh

Lzmin   = 0.
Lzmax   = 5000.
nLz     = 11
Lzgrid  = np.linspace(Lzmin,Lzmax,nLz)

agemin  = 0.
agemax  = taumax
nage    = 100
agegrid = np.linspace(agemin,agemax,nage)

# Define line colors
colors = plt.cm.magma(np.linspace(0,1,nLz))

# Use contourf to provide my colorbar info, then clear
Z = [[0,0],[0,0]]
cf = plt.contourf(Z, Lzgrid, cmap="magma")
plt.clf()

# Create plot
plt.rc('font',family='serif') 
fig = plt.figure(figsize=(7,4))
for jLz in range(nLz):
    Lz  = np.zeros(nage)+Lzgrid[jLz]
    out = cd.ChemicalAbundanceDFSB(fehgrid,Lz,agegrid,taumax,fehbirth,tauenrich,fehLzsolgrad,Lzsol,fehsig)
    fehpred = out[0]
    plt.plot(agegrid,fehpred,color=colors[jLz],linewidth=2.0) 
plt.xlabel(r"$\tau$ (Gyr)",fontsize=16)
plt.ylabel("[Fe/H]",fontsize=16)
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12)
cbar = plt.colorbar(cf)
cbar.ax.tick_params(labelsize=12)
cbar.set_label(r'$L_z$ (km/s kpc)', fontsize=16)
fig.savefig("../plots/edf/feh.eps",format='eps',bbox_inches='tight')
```

## Test surface brightness profile

```python
# Read default parameters
DEFAULTPARSFILE   = "../setupinfo/DoubleDisc/initial_pars_boundaries.csv"
pardict = {}
with open(DEFAULTPARSFILE) as f:
    for line in f:
        (key, val, vallo, valhi, description) = line.split(",")
        pardict[key] = float(val)
        
# Set units for agama library
agama.setUnits(mass=1,length=1,velocity=1)
        
# Create galaxy potential instance
thindisc_par   = dict(type='Disk',
                      surfaceDensity=8.96e+08,#5.70657e+08, 
                      scaleRadius=2.5,#2.6824, 
                      scaleHeight=0.3)#0.1960)
thickdisc_par  = dict(type='Disk', 
                      surfaceDensity=1.83e+08,#2.51034e+08, 
                      scaleRadius=3.0,#2.6824, 
                      scaleHeight=0.9)#0.7010)
HIgasdisc_par  = dict(type='Disk', 
                      surfaceDensity=5.31e+07,#9.45097e+07,
                      scaleRadius=7.,#5.3649, 
                      scaleHeight=-0.085,#0.04,
                      innerCutoffRadius=4.)
HIIgasdisc_par = dict(type='Disk', 
                      surfaceDensity=2.18e+09,
                      scaleRadius=1.5, 
                      scaleHeight=-0.045,
                      innerCutoffRadius=12.)
bulge_par      = dict(type='Spheroid',    
                      densityNorm=9.84e+10,#9.49e+10, 
                      axisRatioZ=0.5, 
                      gamma=0., 
                      beta=1.8,
                      scaleRadius=0.075,
                      outerCutoffRadius=2.1)
dmhalo_par     = dict(type='Spheroid', 
                      densityNorm=pardict["densityNorm_dh"],
                      axisRatioZ=1.0, 
                      gamma=1., 
                      beta=3.,
                      scaleRadius=pardict["scaleRadius_dh"])
galpot_par = [thindisc_par,thickdisc_par,HIgasdisc_par,HIIgasdisc_par,bulge_par,dmhalo_par]
gp         = agama.Potential(galpot_par[0],galpot_par[1],galpot_par[2],galpot_par[3],galpot_par[4],galpot_par[5]) 

# Modify surface brightness parameters
Lz0      = 10.
Rpmin    = 3
Rpmax    = 6
nfallmin = 1.5
nfallmax = 1.5

Lzmin   = 0.
Lzmax   = 4000.
nLz     = 100.
Lzgrid  = np.linspace(Lzmin,Lzmax,int(nLz))

agemin  = 0.
agemax  = 12
nage    = 13
agegrid = np.linspace(agemin,agemax,nage)

# Define line colors
colors = plt.cm.magma(np.linspace(0,1,nage))

# Use contourf to provide my colorbar info, then clear
Z = [[0,0],[0,0]]
cf = plt.contourf(Z, agegrid, cmap="magma")
plt.clf()

# Create plot
plt.rc('font',family='serif') 
fig    = plt.figure(figsize=(7,4))
for jage in range(nage):
    
    # Circular radius and frequencies
    Rc           = gp.Rcirc(L=Lzgrid)
    xc           = np.column_stack((Rc,Rc*0.,Rc*0.))
    force, deriv = gp.forceDeriv(xc) # Returns force and force derivatives at (x, y, z)
    kappa        = np.sqrt(-deriv[:,0] - 3.*force[:,0]/xc[:,0])
    nu           = np.sqrt(-deriv[:,2])
    omega        = np.sqrt(-force[:,0]/xc[:,0]) 
    frequencies  = [kappa,nu,omega]
    prob         = psd.SurfaceBrightnessDFWeibull(Lzgrid,agegrid[jage],Rc,frequencies,Lz0,taumax,Rpmin,Rpmax,nfallmin,nfallmax)
    plt.plot(Rc,prob,color=colors[jage],linewidth=2.0) 

cbar = plt.colorbar(cf)
cbar.set_label(r'$\tau$ (Gyr)', fontsize=16)
    
plt.xlabel(r"$R_c$ (kpc)",fontsize=16)
plt.ylabel("Surface brightness",fontsize=16)
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12)
fig.savefig("../plots/edf/sb.eps",format='eps',bbox_inches='tight')
```

## Test vertical motion distribution function

```python
# Modify radial motion parameters
tausig      = 0.11
R0          = 8.2
sigmaz0tau0 = 45.
Rsigma      = 9.
betaz       = 0.33

# Age
age = 8.

jzmin   = 0. 
jzmax   = 1000.
njz     = 1000
jzgrid  = np.linspace(jzmin,jzmax,njz)
, 
Lzmin   = 0.
Lzmax   = 5000.
nLz     = 11
Lzgrid  = np.linspace(Lzmin,Lzmax,nLz)

# Define line colors
colors = plt.cm.magma(np.linspace(0,1,nLz))

# Use contourf to provide my colorbar info, then clear
Z = [[0,0],[0,0]]
cf = plt.contourf(Z, Lzgrid, cmap="magma")
plt.clf()

# Create plot
plt.rc('font',family='serif') 
fig    = plt.figure(figsize=(7,4))
for jLz in range(nLz):
    
    # Circular radius and frequencies
    Rc           = gp.Rcirc(L=Lzgrid[jLz])
    xc           = np.column_stack((Rc,Rc*0.,Rc*0.))
    force, deriv = gp.forceDeriv(xc) # Returns force and force derivatives at (x, y, z)
    kappa        = np.sqrt(-deriv[:,0] - 3.*force[:,0]/xc[:,0])
    nu           = np.sqrt(-deriv[:,2])
    omega        = np.sqrt(-force[:,0]/xc[:,0]) 
    frequencies  = [kappa,nu,omega]
    prob        = np.zeros(njz)
    for jjz in range(njz):
        prob[jjz] = psd.JrDFExponential(jzgrid[jjz],age,Rc,frequencies,tausig,taumax,R0,sigmaz0tau0,Rsigma,betaz)
    plt.plot(jzgrid,prob,color=colors[jLz],linewidth=2.0) 
    
cbar = plt.colorbar(cf)
cbar.set_label(r'$L_z$ (km/s kpc)', fontsize=16)
    
plt.xscale("log")
plt.xlabel(r"$J_z$ (km/s kpc)",fontsize=16)
plt.ylabel("Probability",fontsize=16)
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12)
fig.savefig("../plots/edf/fjz.eps",format='eps',bbox_inches='tight')
```

## Test heating prescription

```python
R0   = 8.1
taus = 0.11
taum = 10.
def sigma(Lz,tau,sigma0,Rsigma,beta):
    Rc = gp.Rcirc(L=Lz)
    x = sigma0*((tau+taus)/(taum+taus))**beta * np.exp((R0-Rc)/Rsigma)
    return(Rc,x)
```

```python
Lzmin     = 0.
Lzmax     = 5000.
nLz       = 11
Lzgrid    = np.linspace(Lzmin,Lzmax,nLz)

agemin  = 0.
agemax  = 10.
nage    = 11
agegrid = np.linspace(agemin,agemax,nage)

# Define line colors
colors = plt.cm.magma(np.linspace(0,1,nage))

# Use contourf to provide my colorbar info, then clear
Z = [[0,0],[0,0]]
cf = plt.contourf(Z, agegrid, cmap="magma")
plt.clf()

# Create plot
sigma0    = 30.
Rsigma    = 5.
beta      = 1.0
plt.rc('font',family='serif') 
fig    = plt.figure(figsize=(7,4))
for jage in range(nage):
    
    Rcgrid,sigmagrid = sigma(Lzgrid,agegrid[jage],sigma0,Rsigma,beta)
    plt.plot(Rcgrid,sigmagrid,color=colors[jage],linewidth=2.0) 

cbar = plt.colorbar(cf)
cbar.set_label(r'$\tau$ (Gyr)', fontsize=16)
    
plt.xlabel(r"$R_c$ (kpc)",fontsize=16)
plt.ylabel(r"$\sigma$ (km/s)",fontsize=16)
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12)
fig.savefig("../plots/edf/sigma.eps",format='eps',bbox_inches='tight')
```
