# Plot emcee fit

```python
%matplotlib inline
import numpy as np
import emcee
import matplotlib.pyplot as plt
import corner
import pygtc
import pandas as pd
```

## Read emcee output

```python
NMC         = 10                # Number of Monte Carlo samples in error convolution integral
FSTD        = 0.1               # Fraction of initial values of parameters to use as standard deviation of Gaussian ball starting point
NWALKERS    = 28                # Number of walkers in emcee run
SMOVE       = 0.0               # Percentage of move mixture to be sampled with stretch move
DEMOVE      = 0.8               # Percentage of move mixture to be sampled with differential evolution move
DESMOVE     = 0.2               # Percentage of move mixture to be sampled with differential evolution snooker move
DISCSETUP   = "DoubleDisc"      # Configuration of thin disk
DISCSEC     = "all"             # Which section of disk
FILEROOT    = "run12"           # Root of output filenames
SAMPLESFILE = "../results/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
              "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5.h5" 
TRIPLOTFILE = "../plots/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
              "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5_emceecorner.eps" 
CHAINPLOTFILE = "../plots/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
                "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5_emceechain.eps" 
POTENTIALTRIPLOTFILE = "../plots/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
              "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5_emceecorner_potential.eps" 
POTENTIALCHAINPLOTFILE = "../plots/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
                "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5_emceechain_potential.eps" 
INNERDISKTRIPLOTFILE = "../plots/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
              "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5_emceecorner_innerdisk.eps" 
INNERDISKCHAINPLOTFILE = "../plots/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
                "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5_emceechain_innerdisk.eps" 
OUTERDISKTRIPLOTFILE = "../plots/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
              "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5_emceecorner_outerdisk.eps" 
OUTERDISKCHAINPLOTFILE = "../plots/fit/"+DISCSETUP+"/"+DISCSEC+"/"+FILEROOT+"_samples_nmc"+str(NMC)+"_nwalkers"+str(NWALKERS)+"_fstd"+str(FSTD)+\
                "_stretch"+str(SMOVE)+"_de"+str(DEMOVE)+"_des"+str(DESMOVE)+"_fitall_newinitpars5_emceechain_outerdisk.eps" 
```

```python
reader  = emcee.backends.HDFBackend(SAMPLESFILE)
tau     = reader.get_autocorr_time()
```

```python
nwalkers,nfitpars = np.shape(reader)
burnin  = int(2.*np.max(tau))
print("burn-in: {0}".format(burnin))
```

## Plot chains

```python
burnin            = 0
samples           = reader.get_chain(discard=burnin)
log_prob_samples  = reader.get_log_prob(discard=burnin)
log_prior_samples = reader.get_blobs(discard=burnin)
print(np.shape(samples))
```

```python
parnames = np.array([r"$\rho_0$",
                     r"$R_s$",
                     r"$a_{\mathrm{out}}$",
                     r"$\tau_{\mathrm{max,in}}$",
                     r"$\tau_{\mathrm{decay,in}}$",
                     r"$\tau_{\mathrm{enr}}$",
                     r"$g_{\mathrm{dFe},\odot}$",
                     r"$R_{\mathrm{p,min,in}}$",
                     r"$R_{\mathrm{p,max,in}}$",
                     r"$n_{\mathrm{p,min,in}}$",
                     r"$n_{\mathrm{p,max,in}}$",
                     r"$\beta_{z,\mathrm{in}}$",
                     r"$\tau_{\mathrm{max,out}}$",
                     r"$\tau_{\mathrm{decay,out}}$",
                     r"$\mu_{\mathrm{Fe,out}}$",
                     r"$\sigma_{\mathrm{Fe,out}}$",
                     r"$R_{\mathrm{p,min,out}}$",
                     r"$n_{\mathrm{p,min,out}}$",
                     r"$\sigma_{r_0,z,\mathrm{out}}$",
                     r"$\beta_{z,\mathrm{out}}$"])
parids = range(20)
fig, axes = plt.subplots(len(parids), figsize=(6, 24), sharex=True)
fig.subplots_adjust(wspace=0, hspace=0)
for i in range(len(parids)):
    ax = axes[i]
    ax.plot(samples[:,:,parids[i]], "gray",linewidth=0.5)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(parnames[parids[i]])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
fig.savefig(CHAINPLOTFILE,format='eps')
```

## Create confusogram

```python
burnin            = 500
samples           = reader.get_chain(discard=burnin,flat=True)
log_prob_samples  = reader.get_log_prob(discard=burnin,flat=True)
log_prior_samples = reader.get_blobs(discard=burnin,flat=True)
print(np.shape(samples))
```

### All parameters

```python
fig = pygtc.plotGTC(chains=samples,
                    paramNames=parnames,
                    nContourLevels  = 3,
                    plotDensity     = False,
                    smoothingKernel = 1.0)
fig.savefig(TRIPLOTFILE,format='png')
```

### Gravitational potential parameters 

```python
parids = np.array([0,1])
fig = pygtc.plotGTC(chains             = samples[:,parids],
                       paramNames      = parnames[parids],
                       nContourLevels  = 3,
                       plotDensity     = False,
                       smoothingKernel = 1.0,
                       figureSize      = 'MNRAS_column')
fig.savefig(POTENTIALTRIPLOTFILE,format='eps')
```

```python
# Compare with Paul McMillan's results
paulFile    = "../literatureComparisons/MassModelsProductionRunFinal/ProductionRunBig2.tab"
samplesPaul = pd.read_table(paulFile,sep="\s+")
paridsPaul  = ["SphrPar10","SphrPar14"]

nsamples,npars     = np.shape(samples[:,parids])
nsamplesPaul,npars = np.shape(samplesPaul[paridsPaul])

fig = pygtc.plotGTC(chains          = [samples[:,parids],samplesPaul[paridsPaul].to_numpy()],
                    weights         = [np.ones(len(samples)),samplesPaul["N_chain"].to_numpy()],
                    paramNames      = parnames[parids],
                    nContourLevels  = 3,
                    plotDensity     = False,
                    smoothingKernel = 1.0,
                    figureSize      = 'MNRAS_column')
```

```python
# Compare total mass profiles
import agama

def massNFW(r,rho0,rs):
    enclosedMass = 4.*np.pi*rho0 *rs**3*(np.log((rs+r)/rs) + rs/(rs+r) - 1)
    return(enclosedMass)

r200     = 200.
m200     = np.array([massNFW(r200,samples[:,parids[0]],samples[:,parids[1]])])
m200Paul = np.array([massNFW(r200,samplesPaul[paridsPaul[0]].to_numpy(),samplesPaul[paridsPaul[1]].to_numpy())])

fig = pygtc.plotGTC(chains          = [np.transpose(m200)/1.e12,np.transpose(m200Paul)/1.e12],
                    weights         = [np.ones(len(samples)),samplesPaul["N_chain"].to_numpy()],
                    paramNames      = r"$M_{200}/10^{12}M_{\odot}$",
                    nContourLevels  = 3,
                    plotDensity     = False,
                    doOnly1dPlot    = True,
                    smoothingKernel = 1.0,
                    customTickFont  = {'family':'serif', 'size':10},
                    customLabelFont = {'family':'serif', 'size':12},
                    figureSize      = 'MNRAS_column')
```
