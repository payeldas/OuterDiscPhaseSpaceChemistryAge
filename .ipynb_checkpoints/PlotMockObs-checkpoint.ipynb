# Plot mock samples

```python
%matplotlib inline
import numpy as np
import pandas as pd
import CoordTrans as ct
import agama
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from mpl_toolkits.mplot3d import Axes3D
from scipy import stats,interpolate
```

## Read data

```python
datafile   = "../data/LMRC-DR4-VF-SNR30-Gaia.csv"
lamostgaia = pd.read_csv(datafile)

# Collect coordinates (ra/rad, dec/rad, s/kpc, vr/kms-1, mura/masyr-1, 
# mudec/masyr-1, [Fe/H]/dex, [a/Fe]/dex, age/Gyr)
Obs = np.column_stack((lamostgaia["gaiadr2_ra"]/180.*np.pi,
                       lamostgaia["gaiadr2_dec"]/180.*np.pi,
                       lamostgaia["s"],
                       lamostgaia["vr"],
                       lamostgaia["gaiadr2_pmra"],
                       lamostgaia["gaiadr2_pmdec"],
                       lamostgaia["feh"],
                       lamostgaia["afe"],
                       lamostgaia["age"]))

eObs = np.column_stack((lamostgaia["gaiadr2_ra_error"]/180.*np.pi,
                        lamostgaia["gaiadr2_dec_error"]/180.*np.pi,
                        lamostgaia["es"],
                        lamostgaia["evr"],
                        lamostgaia["gaiadr2_pmra_error"],
                        lamostgaia["gaiadr2_pmdec_error"],
                        lamostgaia["efeh"],
                        lamostgaia["eafe"],
                        lamostgaia["eage"]))
print("Read LAMOST DR4 red clump supplemented with Gaia DR2 ...")
                       
# Replace stars without Gaia matches with original sky positions and proper motions
index = (lamostgaia["gaiamatch"]==False) 
Obs[index,0]  = lamostgaia["ra"][index]/180.*np.pi
Obs[index,1]  = lamostgaia["dec"][index]/180.*np.pi
eObs[index,0] = lamostgaia["era"][index]/180.*np.pi
eObs[index,1] = lamostgaia["edec"][index]/180.*np.pi
Obs[index,4]  = lamostgaia["mra"][index]
Obs[index,5]  = lamostgaia["mdec"][index]
eObs[index,4] = lamostgaia["emra"][index]
eObs[index,5] = lamostgaia["emdec"][index]
print("Replaced "+str(np.sum(index))+" stars with missing Gaia data with original data...")

# Replace stars with NaN proper motions with original UCAC5 proper motions
index = (np.isnan(lamostgaia["gaiadr2_pmra"])) | \
        (np.isnan(lamostgaia["gaiadr2_pmdec"])) | \
        (np.isnan(lamostgaia["gaiadr2_pmra_error"])) | \
        (np.isnan(lamostgaia["gaiadr2_pmdec_error"]))
Obs[index,4]  = lamostgaia["mra"][index]
Obs[index,5]  = lamostgaia["mdec"][index]
eObs[index,4] = lamostgaia["emra"][index]
eObs[index,5] = lamostgaia["emdec"][index]
print("Replaced "+str(np.sum(index))+" stars having NaN proper motions with original UCAC5 proper motions...")
```

## Create pure sample of low-alpha stars

```python
# Remove stars that are likely to be massive merged stars
idx_merged_massive = (Obs[:,7]>0.12) & (Obs[:,8]<5.)
print("Removing "+str(np.sum(idx_merged_massive))+" stars that are likelly to be merged stars.")
Obs  = Obs[~idx_merged_massive,:]
eObs = eObs[~idx_merged_massive,:]

# Remove stars with ages older than the age of the Universe
idx_older_universe = Obs[:,8]>13.1
print("Removing a further "+str(np.sum(idx_older_universe))+" stars that are older than the age of the Universe.")
Obs  = Obs[~idx_older_universe,:]
eObs = eObs[~idx_older_universe,:]

# Select low-alpha population
slope      = -0.07
yintercept = 0.12
afemax     = yintercept + slope*Obs[:,6]
idx_lowalpha  = afemax > Obs[:,7]
print("Removing a further "+str(len(Obs) - np.sum(idx_lowalpha))+" stars that belong to the high-alpha disk.")
Obs  = Obs[idx_lowalpha,:]
eObs = eObs[idx_lowalpha,:]
len(Obs)
```

```python
# Number of stars
nstars = len(Obs)

# Convert coordinates to polar
solpos            = np.array([8.2,0.014,-8.6,13.9+232.8,7.1])      
we                = Obs[:,0:6]
wg                = ct.EquatorialToGalactic(we)
wp                = ct.GalacticToPolar(wg,solpos)
wc                = ct.PolarToCartesian(wp)
obsPolarNotRef    = np.column_stack((wp,Obs[:,6],Obs[:,7],Obs[:,8]))
Lz                = obsPolarNotRef[:,0] * obsPolarNotRef[:,4]
```

# Read mock sample with selection function

```python
fileroot    = "test22_nmc1"
samplesfile = "../results/mocksamples/mocksamples_"+fileroot+".txt"
samples     = np.loadtxt(samplesfile)
print(np.shape(samples))

nnoise      = 1
nobs        = 9
samples     = np.reshape(samples,[-1,nobs])
nsamp       = len(samples)/nnoise
fsamp       = nstars/np.float(nsamp)

# Convert coordinates to polar
we                 = samples[:,0:6]
wg                 = ct.EquatorialToGalactic(we)
wp                 = ct.GalacticToPolar(wg,solpos)
samplesPolarNotRef = np.column_stack((wp,samples[:,6],samples[:,7],samples[:,8]))
samplesPolar       = np.copy(samplesPolarNotRef)
samplesPolar[:,2]  = np.abs(samplesPolarNotRef[:,2])

# Reshape accordinate to number of noise samples created
samplesPolar       = np.reshape(samplesPolar,[nsamp,nnoise,nobs])
samplesPolarNotRef = np.reshape(samplesPolarNotRef,[nsamp,nnoise,nobs])
print (np.shape(samplesPolarNotRef))
```

## Plot 2-D $R$-$z$ maps

```python
# 2D density map in R-z
R    = samplesPolarNotRef[:,0,0]
z    = samplesPolarNotRef[:,0,2]
feh  = samplesPolarNotRef[:,0,6]
afe  = samplesPolarNotRef[:,0,7]
age  = samplesPolarNotRef[:,0,8]


plt.rc('font',family='serif')  
fig2 = plt.figure(figsize=(6,4))   
densitymap = plt.hexbin(R,z,bins="log",cmap="jet",gridsize=100,mincnt=1)
plt.xlim([2.,20.])
plt.ylim([-6.,6.])
plt.xlabel(r"$R$ (kpc)",fontsize=16)
plt.ylabel(r"$z$ (kpc)",fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout(pad=1)
plt.colorbar(densitymap).set_label(label=r"$\ln \rho$",size=16)
plotfile = "../plots/mocksamples/2Dhistogram_Rz.eps"
fig2.savefig(plotfile,format='eps')
plotfile = "../plots/mocksamples/2Dhistogram_Rz.png"
fig2.savefig(plotfile,format='png')

# 2D density map in [Fe/H]-z
plt.rc('font',family='serif')
fig3 = plt.figure(figsize=(6,4))   
densitymap = plt.hexbin(feh,z,bins="log",cmap="jet",gridsize=60,mincnt=1)
plt.xlim([-1.0,0.5])
plt.ylim([-6.,6.])
plt.xlabel("[Fe/H] (dex)",fontsize=16)
plt.ylabel(r"$z$ (kpc)",fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout(pad=1)
plt.colorbar(densitymap).set_label(label=r"$\ln \rho$",size=16)
plotfile = "../plots/mocksamples/2Dhistogram_fehz.eps"
fig3.savefig(plotfile,format='eps')
plotfile = "../plots/mocksamples/2Dhistogram_fehz.png"
fig3.savefig(plotfile,format='png')

# 2D density map in [a/Fe]-z
plt.rc('font',family='serif')
fig3 = plt.figure(figsize=(6,4))   
densitymap = plt.hexbin(afe,z,bins="log",cmap="jet",gridsize=60,mincnt=1)
plt.xlim([-0.2,0.45])
plt.ylim([-6.,6.])
plt.xlabel(r"[$\alpha$/Fe] (dex)",fontsize=16)
plt.ylabel(r"$z$ (kpc)",fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout(pad=1)
plt.colorbar(densitymap).set_label(label=r"$\ln \rho$",size=16)
plotfile = "../plots/mocksamples/2Dhistogram_afez.eps"
fig3.savefig(plotfile,format='eps')
plotfile = "../plots/mocksamples/2Dhistogram_afez.png"
fig3.savefig(plotfile,format='eps')

# 2D age map in R-z
plt.rc('font',family='serif')  
fig4 = plt.figure(figsize=(6,4))   
agemap = plt.hexbin(R,z,C=age,cmap="jet",gridsize=60,mincnt=1)
plt.xlim([2.,20.])
plt.ylim([-6.,6.])
plt.xlabel(r"$R$ (kpc)",fontsize=16)
plt.ylabel(r"$z$ (kpc)",fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout(pad=1)
plt.colorbar(agemap).set_label(label=r"$\tau$ (Gyr)",size=16)
plotfile = "../plots/mocksamples/2Dhistogram_Rz_age.eps"
fig4.savefig(plotfile,format='eps')

# 2D sigma_z map in R-z
vz = obsPolarNotRef[:,5]
plt.rc('font',family='serif')  
fig5 = plt.figure(figsize=(6,4))   
dispmap = plt.hexbin(R,z,C=vz,reduce_C_function=np.std,vmin=0.,vmax=60.,
                     cmap="jet",gridsize=60,mincnt=1)
plt.xlim([2.,20.])
plt.ylim([-6.,6.])
plt.xlabel(r"$R$ (kpc)",fontsize=16)
plt.ylabel(r"$z$ (kpc)",fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout(pad=1)
plt.colorbar(dispmap).set_label(label=r"$\sigma_z$ (km/s)",size=16)
plotfile = "../plots/mocksamples/2Dhistogram_Rz_sigz.eps"
fig5.savefig(plotfile,format='eps')
plotfile = "../plots/mocksamples/2Dhistogram_Rz_sigz.png"
fig5.savefig(plotfile,format='png')
```
