## Read data

```python
datafile   = "../data/LMRC-DR4-VF-SNR30-newnames-GaiaeDR3.csv"
lamostgaia = pd.read_csv(datafile)

solpos = np.array([8.2,0.014,-8.6,13.9+232.8,7.1]) # McMillan 2017

# Collect coordinates (ra/rad, dec/rad, s/kpc, vr/kms-1, mura/masyr-1, 
# mudec/masyr-1, [Fe/H]/dex, [a/Fe]/dex, age/Gyr)
Obs = np.column_stack((lamostgaia["ra"]/180.*np.pi,
                       lamostgaia["dec"]/180.*np.pi,
                       lamostgaia["s"],
                       lamostgaia["vr"],
                       lamostgaia["pmra"],
                       lamostgaia["pmdec"],
                       lamostgaia["feh"],
                       lamostgaia["afe"],
                       lamostgaia["age"]))

eObs = np.column_stack((lamostgaia["ra_error"]/180.*np.pi,
                        lamostgaia["dec_error"]/180.*np.pi,
                        lamostgaia["es"],
                        lamostgaia["evr"],
                        lamostgaia["pmra_error"],
                        lamostgaia["pmdec_error"],
                        lamostgaia["efeh"],
                        lamostgaia["eafe"],
                        lamostgaia["eage"]))

feh = np.copy(lamostgaia["feh"])
afe = np.copy(lamostgaia["afe"])

print("Read LAMOST DR4 red clump supplemented with Gaia eDR3 ...")
                       
# Replace stars with NaN proper motions with original UCAC5 proper motions
index = (np.isnan(lamostgaia["pmra"])) | \
        (np.isnan(lamostgaia["pmdec"])) | \
        (np.isnan(lamostgaia["pmra_error"])) | \
        (np.isnan(lamostgaia["pmdec_error"]))
Obs[index,4]  = lamostgaia["mra"][index]
Obs[index,5]  = lamostgaia["mdec"][index]
eObs[index,4] = lamostgaia["emra"][index]
eObs[index,5] = lamostgaia["emdec"][index]
print("Replaced "+str(np.sum(index))+" stars having NaN proper motions with original UCAC5 proper motions...")
```

## Create pure sample of low-alpha stars

```python
# Remove stars that are likely to be massive merged stars
idx_merged_massive = (afe>0.12) & (Obs[:,8]<5.)
print("Removing "+str(np.sum(idx_merged_massive))+" stars that are likelly to be merged stars.")
Obs  = Obs[~idx_merged_massive,:]
eObs = eObs[~idx_merged_massive,:]
feh  = feh[~idx_merged_massive]
afe  = afe[~idx_merged_massive]

# Remove stars with ages older than the age of the Universe
idx_older_universe = Obs[:,8]>13.1
print("Removing a further "+str(np.sum(idx_older_universe))+" stars that are older than the age of the Universe.")
Obs  = Obs[~idx_older_universe,:]
eObs = eObs[~idx_older_universe,:]
feh  = feh[~idx_older_universe]
afe  = afe[~idx_older_universe]

# Select low-alpha population
slope         = -0.07
yintercept    = 0.12
afemax        = yintercept + slope*feh
idx_lowalpha  = afemax > afe
print("Removing a further "+str(len(feh) - np.sum(idx_lowalpha))+" stars that belong to the high-alpha disc.")
Obs  = Obs[idx_lowalpha,:]
eObs = eObs[idx_lowalpha,:]
feh  = feh[idx_lowalpha]
afe  = afe[idx_lowalpha]

# Number of stars
nstars = len(afe)
print(nstars)
```

```python
# Convert coordinates to polar coordinates with errors
nmc    = 100
solpos = np.array([8.2,0.014,-8.6,13.9+232.8,7.1])      

ra     = Obs[:,0]
dec    = Obs[:,1]
s      = Obs[:,2]
vr     = Obs[:,3]
mura   = Obs[:,4]
mudec  = Obs[:,5]
era    = eObs[:,0]
edec   = eObs[:,1]
es     = eObs[:,2]
mulns  = np.log(s/np.sqrt(1.+es**2./s**2.))
siglns = np.sqrt(np.log(1.+es**2./s**2.))
evr    = eObs[:,3]
emura  = eObs[:,4]
emudec = eObs[:,5]
```

```python
# Generate Monte Carlo samples
ramc    = np.random.normal(ra,era,size=[nmc,nstars])
decmc   = np.random.normal(dec,edec,size=[nmc,nstars])
smc     = np.random.lognormal(mulns,siglns,size=[nmc,nstars])
vrmc    = np.random.normal(vr,evr,size=[nmc,nstars])
muramc  = np.random.normal(mura,emura,size=[nmc,nstars])
mudecmc = np.random.normal(mudec,emudec,size=[nmc,nstars])
```

```python
# Convert to polar coordinates
wpmc = np.zeros([nmc,nstars,6])
for jmc in range(nmc):
    we = np.column_stack((ramc[jmc,:],decmc[jmc,:],smc[jmc,:],
                          vrmc[jmc,:],muramc[jmc,:],mudecmc[jmc,:]))
    wg            = ct.EquatorialToGalactic(we)
    wpmc[jmc,:,:] = ct.GalacticToPolar(wg,solpos)
    
obsPolarNotRef = np.column_stack((np.mean(wpmc,0),Obs[:,6],Obs[:,7],Obs[:,8]))
obsPolar       = np.copy(obsPolarNotRef)
obsPolar[:,2]  = np.abs(obsPolar[:,2])
eObsPolar      = np.column_stack((np.std(wpmc,0),eObs[:,6],eObs[:,7],Obs[:,8]))
```

## Read mock samples with selection function

```python
samples     = np.loadtxt(MOCKOBSFILE)
nnoise      = 1
nobs        = 9
samples     = np.reshape(samples,[-1,nobs])
nsamp       = int(len(samples)/nnoise) 
fsamp       = nstars/np.float(nsamp)

# Convert coordinates to polar
we                 = samples[:,0:6]
wg                 = ct.EquatorialToGalactic(we)
wp                 = ct.GalacticToPolar(wg,solpos)
samplesPolarNotRef = np.column_stack((wp,samples[:,6],samples[:,7],samples[:,8]))
samplesPolar       = np.copy(samplesPolarNotRef)
samplesPolar[:,2]  = np.abs(samplesPolarNotRef[:,2])

# Reshape according to number of noise samples created
samples            = np.reshape(samples,[nsamp,nnoise,nobs])
samplesPolar       = np.reshape(samplesPolar,[nsamp,nnoise,nobs])
samplesPolarNotRef = np.reshape(samplesPolarNotRef,[nsamp,nnoise,nobs])
```

## Read mock samples without selection function

```python
trueSamples = np.loadtxt(MOCKOBSFILE_NOSF)
nnoise_nosf = 1
nobs        = 9
trueSamples = np.reshape(trueSamples,[-1,nobs])
nsamp_nosf  = int(len(trueSamples)/nnoise_nosf)
fsamp_nosf  = nstars/np.float(nsamp_nosf)

# Convert coordinates to polar
we              = trueSamples[:,0:6]
wg              = ct.EquatorialToGalactic(we)
wp              = ct.GalacticToPolar(wg,solpos)
truePolarNotRef = np.column_stack((wp,trueSamples[:,6],trueSamples[:,7],trueSamples[:,8]))
truePolar       = np.copy(truePolarNotRef)
truePolar[:,2]  = np.abs(truePolarNotRef[:,2])

# Reshape accordinate to number of noise samples created
trueSamples     = np.reshape(trueSamples,[nsamp,nnoise_nosf,nobs])
truePolar       = np.reshape(truePolar,[nsamp,nnoise_nosf,nobs])
truePolarNotRef = np.reshape(truePolarNotRef,[nsamp,nnoise_nosf,nobs])
```

```python
median     = 50.
onesig_lo  = median-34.1
onesig_hi  = median+34.1
twosig_lo  = median-47.7
twosig_hi  = median+47.7

Rmin   = solpos[0]-0.2
Rmax   = solpos[0]+0.2
nz     = 12
zbins  = np.linspace(0.,ZMAX,nz)

obsRz     = np.column_stack((obsPolar[:,0],obsPolar[:,2]))
obskz     = np.zeros(nz)
sampleskz = np.zeros([nnoise,nz])
sampleskz_med = np.zeros(nz)
sampleskz_onesig_lo = np.zeros(nz)
sampleskz_onesig_hi = np.zeros(nz)
sampleskz_twosig_lo = np.zeros(nz)
sampleskz_twosig_hi = np.zeros(nz)
truekz = np.zeros([nnoise_nosf,nz])
truekz_med = np.zeros(nz)
truekz_onesig_lo = np.zeros(nz)
truekz_onesig_hi = np.zeros(nz)
truekz_twosig_lo = np.zeros(nz)
truekz_twosig_hi = np.zeros(nz)
for jz in range(nz):
    index         = obsRz[:,1]<zbins[jz]
    obskz[jz]     = np.sum(index)
    for jnoise in range(nnoise):
        index  = samplesPolar[:,jnoise,2]<zbins[jz]
        sampleskz[jnoise,jz] = np.sum(index)  
    sampleskz_med[jz] = np.percentile(sampleskz[:,jz],median)
    sampleskz_onesig_lo[jz] = np.percentile(sampleskz[:,jz],onesig_lo)
    sampleskz_onesig_hi[jz] = np.percentile(sampleskz[:,jz],onesig_hi)
    sampleskz_twosig_lo[jz] = np.percentile(sampleskz[:,jz],twosig_lo)
    sampleskz_twosig_hi[jz] = np.percentile(sampleskz[:,jz],twosig_hi)
    for jnoise in range(nnoise_nosf):
        index  = truePolar[:,jnoise,2]<zbins[jz]
        truekz[jnoise,jz] = np.sum(index)  
    truekz_med[jz] = np.percentile(truekz[:,jz],median)
    truekz_onesig_lo[jz] = np.percentile(truekz[:,jz],onesig_lo)
    truekz_onesig_hi[jz] = np.percentile(truekz[:,jz],onesig_hi)
    truekz_twosig_lo[jz] = np.percentile(truekz[:,jz],twosig_lo)
    truekz_twosig_hi[jz] = np.percentile(truekz[:,jz],twosig_hi)
fig = plt.figure(figsize=(6,4)) 
plt.rcParams['text.usetex'] = True
plt.tick_params(direction="in")
plt.plot(zbins,obskz,'-ok',ms=6,linewidth=2,zorder=200,label="Obs")
plt.plot(zbins,np.transpose(sampleskz),'-',color="steelblue",linewidth=2,label='Mock obs')
plt.plot(zbins,np.transpose(truekz),'-',color='lightskyblue',linewidth=2,label='Mock eDF')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel(r"$|z|$ (kpc)",fontsize=14)
plt.ylabel(r"No. stars within height $z$ at $R_0$ ($\sim K_z$)",fontsize=14)
#plt.legend()
plotfile = PLOTSDIR+"verticalkz.eps"
fig.savefig(plotfile,format='eps')
```

## Kinematic profiles

### 1. 1D robust vertical velocity dispersion profiles below and above plane by age

```python
# Define objective function to maximize for estimation of robust velocity dispersion
# sigma - velocity dispersion
# vel   - vector of velocities
# eVel  - vector of velocity uncertainties
# mu    - mean velocity
def sigmaObjFunc(sigma,vel,eVel,mu):
    
    y = ((vel-mu)**2.)/((sigma**2.+eVel**2.)**2.) - (1./(sigma**2.+eVel**2.))
    
    return (np.sum(y))

# Number of sigma for clipping
f = 5.
```

```python
Rmin      = 2.
Rmax      = 24.
Redges    = np.array([Rmin,7.,7.5,8.,8.5,9.0,9.5,10.,10.5,11.,12.,13.,Rmax])
nR        = len(Redges)-1
zmin      = 0.6
zmax      = 1.0
agemin    = np.min(obsPolar[:,8])
agemax    = np.max(obsPolar[:,8])
ageedges  = np.array([agemin,2.,3.,4.,5.,6.,7.,8.,agemax])
nage      = len(ageedges)-1

obsR          = np.zeros(nR)
obsSigmaz     = np.zeros(nR)
obsSigmaz_err = np.zeros(nR)
sampR         = np.zeros(nR)
sampSigmaz    = np.zeros(nR)
trueR         = np.zeros(nR)
trueSigmaz    = np.zeros(nR)
binned_age    = np.zeros(nage)

# Plot each line with a different colour
c = np.arange(1, nage + 1)
plot_agemin = 1.5
plot_agemax = 9.5
norm = mpl.colors.Normalize(vmin=plot_agemin, vmax=plot_agemax)
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.magma)
cmap.set_array([]) 

# Number of sigma for clipping
f = 5.

# z range
zmin = 0.6
zmax = 1.   

# Dispersion minimum and maximum
intervalMin = 0.
intervalMax = 100.
        
fig,ax = plt.subplots(3,1,figsize=(7,12))  
plt.rcParams['text.usetex'] = True
plt.tick_params(direction="in")
for jage in range(nage):
               
    for jR in range(nR):
        
        ## OBSERVATIONS
        # Define Rz-age box
        boxIndex = ((obsPolar[:,2]>zmin) & (obsPolar[:,2]<zmax) &\
                    (Redges[jR]<obsPolar[:,0]) & (obsPolar[:,0]<Redges[jR+1]) &\
                    (ageedges[jage]<obsPolar[:,8]) & (obsPolar[:,8]<ageedges[jage+1]))
               
        # Calculate mean velocity
        mean = np.mean(obsPolar[boxIndex,5])
        
        # Calculate initial guess for velocity dispersion
        sigma = np.std(obsPolar[boxIndex,5])
        
        # Clip velocities beyond three sigma
        clippedIndex = (obsPolar[:,5] > (mean-f*sigma)) &\
                       (obsPolar[:,5] < (mean+f*sigma))
        index        = clippedIndex * boxIndex
        
        # Calculate mean radii of clipped velocities
        obsR[jR] = np.mean(obsPolar[index,0])
        
        n = np.sum(index)
        
        # Mean age
        ageIndex = (ageedges[jage]<obsPolar[:,8]) & (obsPolar[:,8]<ageedges[jage+1])
        binned_age[jage] = np.mean(obsPolar[ageIndex,8])
        
        # Recalculate mean
        mean = np.mean(obsPolar[index,5])
        
        # Maximise likelihood to find robust measure of observed velocity dispersion
        optSigma = sciOpt.brentq(sigmaObjFunc,intervalMin,intervalMax, 
                                 args=(obsPolar[index,5],eObsPolar[index,5],mean), 
                                 xtol=2e-12, rtol=8.881784197001252e-16, 
                                 maxiter=100, full_output=False)
        obsSigmaz[jR] = optSigma
        n = np.sum(index)
        obsSigmaz_err[jR] = optSigma * np.sqrt(1./(2.*n))
        
        ## MOCK OBSERVATIONS
        # Define Rz-age box
        boxIndex = ((samplesPolar[:,0,2]>zmin) & (samplesPolar[:,0,2]<zmax) &\
                    (Redges[jR]<samplesPolar[:,0,0]) & (samplesPolar[:,0,0]<Redges[jR+1]) &\
                    (ageedges[jage]<samplesPolar[:,0,8]) & (samplesPolar[:,0,8]<ageedges[jage+1]))
               
        # Calculate mean velocity
        mean = np.mean(samplesPolar[boxIndex,0,5])
        
        # Calculate initial guess for velocity dispersion
        sigma = np.std(samplesPolar[boxIndex,0,5])
        
        # Clip velocities beyond three sigma
        clippedIndex = (samplesPolar[:,0,5] > (mean-f*sigma)) &\
                       (samplesPolar[:,0,5] < (mean+f*sigma))
        index        = clippedIndex * boxIndex
        
        # Calculate mean radii of clipped velocities
        sampR[jR] = np.mean(samplesPolar[index,0,0])
                       
        # Recalculate mean
        mean = np.mean(samplesPolar[index,0,5])
        
        # Maximise likelihood to find robust measure of observed velocity dispersion
        optSigma = sciOpt.brentq(sigmaObjFunc,intervalMin,intervalMax, 
                                 args=(samplesPolar[index,0,5],samplesPolar[index,0,5]*0.+0.5,mean), 
                                 xtol=2e-12, rtol=8.881784197001252e-16, 
                                 maxiter=100, full_output=False)
        sampSigmaz[jR] = optSigma
        
        ## TRUE
        # Define Rz-age box
        boxIndex = ((truePolar[:,0,2]>zmin) & (truePolar[:,0,2]<zmax) &\
                    (Redges[jR]<truePolar[:,0,0]) & (truePolar[:,0,0]<Redges[jR+1]) &\
                    (ageedges[jage]<truePolar[:,0,8]) & (truePolar[:,0,8]<ageedges[jage+1]))
               
        # Calculate mean velocity
        mean = np.mean(truePolar[boxIndex,:,5])
        
        # Calculate initial guess for velocity dispersion
        sigma = np.std(truePolar[boxIndex,:,5])
        
        # Clip velocities beyond three sigma
        clippedIndex = (truePolar[:,0,5] > (mean-f*sigma)) &\
                       (truePolar[:,0,5] < (mean+f*sigma))
        index        = clippedIndex * boxIndex
        
        # Calculate mean radii of clipped velocities
        trueR[jR] = np.mean(truePolar[index,0,0])
                
        # Recalculate mean
        mean = np.mean(truePolar[index,0,5])
        
        # Maximise likelihood to find robust measure of observed velocity dispersion

        optSigma = sciOpt.brentq(sigmaObjFunc,intervalMin,intervalMax, 
                                 args=(truePolar[index,0,5],truePolar[index,0,5]*0.+0.5,mean), 
                                 xtol=2e-12, rtol=8.881784197001252e-16, 
                                 maxiter=100, full_output=False)
        trueSigmaz[jR] = optSigma
               
    pl=ax[0].errorbar(obsR,obsSigmaz,yerr=obsSigmaz_err,
                   linestyle="-",linewidth=2,fmt="o",color=cmap.to_rgba(jage + 1),ms=2,
                   label="Data",zorder=200)
    pl=ax[1].errorbar(sampR,sampSigmaz,linestyle="-",linewidth=2,fmt="o",color=cmap.to_rgba(jage + 1),ms=2,
                   label="Data",zorder=200)
    pl=ax[2].errorbar(trueR,trueSigmaz,linestyle="-",linewidth=2,fmt="o",color=cmap.to_rgba(jage + 1),ms=2,
                   label="Data",zorder=200)
    
ax[0].tick_params(axis="x",labelsize=12)
ax[0].tick_params(axis="y",labelsize=12)
ax[0].set_xlabel(r"$R$ (kpc)",fontsize=16)
ax[0].set_ylabel(r"$\sigma_z$ (km/s)",fontsize=16)
ax[0].set_title(r"$0.6<|z|<1.0$ kpc")
ax[0].set_xlim([obsR[0],obsR[nR-1]])
ax[0].set_ylim([10.,40.])

ax[1].tick_params(axis="x",labelsize=12)
ax[1].tick_params(axis="y",labelsize=12)
ax[1].set_xlabel(r"$R$ (kpc)",fontsize=16)
ax[1].set_ylabel(r"$\sigma_z$ (km/s)",fontsize=16)
ax[1].set_title(r"$0.6<|z|<1.0$ kpc")
ax[1].set_xlim([obsR[0],obsR[nR-1]])
ax[1].set_ylim([10.,40.])

ax[2].tick_params(axis="x",labelsize=12)
ax[2].tick_params(axis="y",labelsize=12)
ax[2].set_xlabel(r"$R$ (kpc)",fontsize=16)
ax[2].set_ylabel(r"$\sigma_z$ (km/s)",fontsize=16)
ax[2].set_title(r"$0.6<|z|<1.0$ kpc")
ax[2].set_xlim([obsR[0],obsR[nR-1]])
ax[2].set_ylim([10.,40.])
    
plt.tight_layout(pad=1)
ticks_min = int(np.floor(np.min(binned_age)))
ticks_max = int(np.ceil(np.max(binned_age)))
cbar=plt.colorbar(cmap,ax=ax,ticks=range(ticks_min,ticks_max))
cbar.set_label(r"$\tau$ (Gyr)",size=16)

plotfile = PLOTSDIR+"robust_sigz_byR_byage.eps"
fig.savefig(plotfile,format='eps')
```

```python
Redges    = np.array([RMIN,8.,10.,12.,RMAX])
nR        = len(Redges)-1
zedges    = np.array([0.,0.5,1.0,ZMAX])
nz        = len(zedges)-1
ageedges  = np.array([AGEMIN,2.,5.,8.,11.,AGEMAX])
nage      = len(ageedges)-1

obsRzage            = np.column_stack((obsPolar[:,0],obsPolar[:,2],obsPolar[:,8]))
obsbinnedRzage,tmp  = np.histogramdd(obsRzage,bins=[Redges,zedges,ageedges]) 
sampbinnedRzage     = np.zeros([nnoise,nR,nz,nage])
truebinnedRzage     = np.zeros([nnoise_nosf,nR,nz,nage])

obsAge        = np.zeros(nage)
obsSigmaz     = np.zeros(nage)
obsSigmaz_err = np.zeros(nage)
sampAge       = np.zeros([nnoise,nage])
sampSigmaz    = np.zeros([nnoise,nage])
trueAge       = np.zeros([nnoise_nosf,nage])
trueSigmaz    = np.zeros([nnoise_nosf,nage])

fig,axarr = plt.subplots(nR,nz,figsize=(9,8),sharex=True,sharey=True)   
plt.rcParams['text.usetex'] = True
plt.tick_params(direction="in")
for jR in range(nR):
    for jz in range(nz):
        for jage in range(nage):
            index                 = ((Redges[jR]<obsPolar[:,0]) &\
                                     (obsPolar[:,0]<Redges[jR+1]) &\
                                     (zedges[jz] < obsPolar[:,2]) &\
                                     (obsPolar[:,2]< zedges[jz+1]) &\
                                     (ageedges[jage]<obsPolar[:,8]) &\
                                     (obsPolar[:,8]<ageedges[jage+1]))
            obsAge[jage]    = np.mean(obsPolar[index,8])
            obsSigmaz[jage] = np.std(obsPolar[index,5])
            n = np.sum(index)
            obsSigmaz_err[jage] = obsSigmaz[jage] * np.sqrt(1./(2.*n))
            for jnoise in range(nnoise):
                index              = (Redges[jR]<samplesPolar[:,jnoise,0]) &\
                                     (samplesPolar[:,jnoise,0]<Redges[jR+1]) &\
                                     (zedges[jz] < samplesPolar[:,jnoise,2]) &\
                                     (samplesPolar[:,jnoise,2]< zedges[jz+1]) &\
                                     (ageedges[jage]<samplesPolar[:,jnoise,8]) &\
                                     (samplesPolar[:,jnoise,8]<ageedges[jage+1])
                sampAge[jnoise,jage]    = np.mean(samplesPolar[index,jnoise,8])
                sampSigmaz[jnoise,jage] = np.std(samplesPolar[index,jnoise,5])
            for jnoise in range(nnoise_nosf):
                index              = (Redges[jR]<truePolar[:,jnoise,0]) &\
                                     (truePolar[:,jnoise,0]<Redges[jR+1]) &\
                                     (zedges[jz] < truePolar[:,jnoise,2]) &\
                                     (truePolar[:,jnoise,2]< zedges[jz+1]) &\
                                     (ageedges[jage]<truePolar[:,jnoise,8]) &\
                                     (truePolar[:,jnoise,8]<ageedges[jage+1])
                trueAge[jnoise,jage]    = np.mean(truePolar[index,jnoise,8])
                trueSigmaz[jnoise,jage] = np.std(truePolar[index,jnoise,5])
        axarr[jR,jz].errorbar(obsAge,obsSigmaz,yerr=obsSigmaz_err,
                         linestyle="-",fmt="o",color="k",ms=6,label="Obs",
                         zorder=200)
        for jnoise in range(nnoise):
            if (jnoise==0):
                axarr[jR,jz].plot(sampAge[jnoise,:],sampSigmaz[jnoise,:],'-',
                                 color="steelblue",label="Mock obs")
            axarr[jR,jz].plot(sampAge[jnoise,:],sampSigmaz[jnoise,:],'-',
                             color="steelblue")
        for jnoise in range(nnoise_nosf):
            if (jnoise==0):
                axarr[jR,jz].plot(trueAge[jnoise,:],trueSigmaz[jnoise,:],'-',
                                 color="lightskyblue",label="Mock eDF")
            axarr[jR,jz].plot(trueAge[jnoise,:],trueSigmaz[jnoise,:],'-',
                             color="lightskyblue")
        if (jR==nR-1):
            axarr[jR,jz].set_xlabel(str(np.round(zedges[jz],1))+r"$<|z|<$"+str(np.round(zedges[jz+1],1))+" kpc",fontsize=10)
        if (jz==0):
            axarr[jR,jz].set_ylabel(str(np.round(Redges[jR],1))+r"$<R<$"+str(np.round(Redges[jR+1],1))+" kpc",fontsize=10)
axarr[jR,0].set_ylim(0.,60.)
axarr[0,0].legend(loc="upper left") 
plt.gcf().text(0.03,0.45,r"$\sigma_z$ (km/s)",fontsize=14,rotation=90)
plt.setp([a.get_xticklabels() for a in axarr[0, :]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False)
plt.gcf().text(0.5,0.01,r"$\tau$ (Gyr)",fontsize=14)
fig.subplots_adjust(hspace=0,wspace=0)
plotfile = PLOTSDIR+"avd_byR_byz.eps"
fig.savefig(plotfile,format='eps')
```
